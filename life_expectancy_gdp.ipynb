## Analysis On GDP And Life Expectancy 

In this project i will analyse on GDP and life expectancy based on data from the World Health Organization and the World Bank. I will try identify the relationship between the GDP and life expectancy of six countries.

### Step 0: Import necessary libraries


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
## import csv data 
df = pd.read_csv('all_data.csv')
print(df.head())
```

```python
print(df.info())
```

```python
df.Country.nunique()
```

```python
df.Year.nunique()
```

```python
df.isnull().sum()
```

```python
unique_countries = df.Country.unique()
```

```python
print(unique_countries)
```

```python
df['GDP_in_bil']= round(df.GDP/1000000000,3)
```

```python
for country in unique_countries:
    sns.lineplot(x= 'Year', y= 'GDP_in_bil', data = df[df.Country== country], color= 'blue', marker='o')
    plt.title(country)
    plt.xticks(rotation=30)
    plt.ylabel('GDP(billion)')
    plt.show()
    plt.clf()
                

```

```python
sns.scatterplot(y='GDP_in_bil', x= 'Life expectancy at birth (years)', data=df, hue=df.Country)
plt.yscale('log')
plt.show()
plt.clf()
```

```python
for country in unique_countries:
    sns.lmplot(x='GDP_in_bil', y= 'Life expectancy at birth (years)', data=df[df.Country == country])
    plt.title(f'{country} GDP vs Life Expectancy')
    plt.show()
    plt.clf()
    
```

## Summary

Based on analysis and visualization, we can conclude there is correlation between GDP and life expectancy.  This may be because of the goof healthcare,access to food ensure higher life expectancy.
